```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from time import time

from sklearn.cluster import KMeans
```

```python
def brush(filename, K=3, aspect=None, fit=True):
    image = Image.open(filename)
    aspect = aspect or 1.
    new_width = int(aspect * image.width)
    new_height = int(aspect * image.height)
    image = image.resize((new_width, new_height))

    image_arr = np.array(image, dtype='float32')
    if image_arr.shape[-1] == 4:
        image_arr = image_arr[:, :, :-1]
    if not fit:
        return np.uint8(image_arr)
    image_shape = image_arr.shape
    image_arr = image_arr.reshape(-1, 3)
    
    kmeans = KMeans(n_clusters=K, random_state=0)
    kmeans.fit(image_arr)
    clusters = kmeans.predict(image_arr)
    image_final = kmeans.cluster_centers_[clusters]
    
    image_final = np.uint8(image_final) 
    image_final = image_final.reshape(image_shape)
    image_arr = np.uint8(image_arr)
    image_arr = image_arr.reshape(image_shape)
    
    return image_arr, image_final, kmeans

def sideplot(image, result, tag=('original', 'result'), figsize=(8, 4), title=None):
    fig, ax = plt.subplots(1, 2, figsize=figsize)
    ax[0].imshow(image)
    ax[0].axis('off')
    ax[0].set_title(tag[0])
    ax[1].imshow(result)
    ax[1].axis('off')
    ax[1].set_title(tag[1])
    fig.tight_layout()
    if title:
        fig.suptitle(title)
    plt.show()
    
def fourplot(image1, image2, result1, image_cross):
    fig, ax = plt.subplots(2, 2, figsize=(8, 8))

    ax[0][0].imshow(image1)
    ax[0][0].set_title('Image 1')
    ax[0][0].axis('off')

    ax[0][1].imshow(image2)
    ax[0][1].set_title('Image 2')
    ax[0][1].axis('off')

    ax[1][0].imshow(result1)
    ax[1][0].set_title('Brushed Image 1')
    ax[1][0].axis('off')

    ax[1][1].imshow(image_cross)
    ax[1][1].set_title('Transferred Palette')
    ax[1][1].axis('off')

    fig.tight_layout()
    plt.show()
    
def transfer_palette(filename1, filename2, K, aspect=None):
    image1, result1, algo1 = brush(filename1, K=K, aspect=aspect)
    image2  = brush(filename2, K=K, aspect=aspect, fit=False)

    orig_shape = image2.shape
    image_cross = image2.reshape(-1, image2.shape[-1])
    clusters = algo1.predict(image_cross)
    image_cross = algo1.cluster_centers_[clusters]
    image_cross = np.uint8(image_cross.reshape(orig_shape))
    
    return (
        image_cross,
        image1,
        result1,
        algo1,
        image2,
        result2,
        algo2
    )
```

```python
filename1, filename2 = 'images/bali.png', 'images/blossom.jpg'
K = 5
aspect = 0.5

image_cross, image1, result1, algo1, image2, result2, algo2 = transfer_palette(filename1, filename2, K=K, aspect=aspect)
fourplot(image1, image2, result1, image_cross)
```

```python
filename1, filename2 = 'images/street5.jpg', 'images/house.jpg'
K = 7
aspect = 0.5

image_cross, image1, result1, algo1, image2, result2, algo2 = transfer_palette(filename1,
                                                                               filename2,
                                                                               K=K, aspect=aspect)
fourplot(image1, image2, result1, image_cross)
```
